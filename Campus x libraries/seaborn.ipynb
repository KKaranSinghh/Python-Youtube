```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly. express as px
```

```python
tips= sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axis level function
sns.scatterplot(data= tips, x= 'total_bill', y= 'tip', hue='sex')
```

```python
# relplot -> figure level -> square shape
sns.relplot(data=tips, x= 'total_bill', y= 'tip', kind= 'scatter', hue= 'sex',style= 'time',size= 'size')
```

### 2. Distribution Plots
- Used for univariate analysis
- Used to find out the distribution
- Range of the observation
- Central Tendency
- Is the data bimodal?
- Are there outlines?
### Plots under distribution plot
- histplot
- kdeplot
- rugplot

```python
# figure level function -> displot
# axis level funciton -> histplot, kdeplot, rugplot
```

```python
tips  # (data)
```

```python
# ploting univariate histogram
sns.histplot(data=tips, x= 'total_bill')
```

```python
# displot
sns.displot(data=tips, x= 'total_bill', kind= 'hist')
```

```python
# bins parameter
sns.displot(data=tips, x= 'total_bill', kind= 'hist',bins=20)
```

```python
# it's also possible to visualize the distribution of a categorical variable using the logic of a histogram.
# discrete bins are automatically set for categorical variables

# histogram acting like a countplot
sns.displot(data=tips, x= 'day', kind= 'hist')
```

```python
# hue parameter
sns.displot(data=tips, x= 'tip', kind= 'hist',hue= 'sex')
```

```python
# element = step  ( it shows lining of the hidden bars)
sns.displot(data=tips, x= 'tip', kind= 'hist',hue= 'sex',element='step')
```

```python
titanic= sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x= 'age', kind= 'hist',hue= 'sex',element='step')
```

```python
# faceting using col and row -> not work on histplot function
sns.displot(data=tips, x= 'tip', kind= 'hist',col= 'sex',element='step')
```

```python
# kdeplot (kernel density estimation)
# Rather then using discrete bins, a KDE plot smooths the observations with a gaussian kernel, producing a continuous density estimate
sns.kdeplot(data= tips, x= 'total_bill')
```

```python
 # same graph using displot
sns.displot(data= tips, x= 'total_bill', kind= 'kde')
```

```python
# hue , fill
sns.displot(data= tips, x= 'total_bill', kind= 'kde',hue='sex',fill=True)
```

```python
# Rugplot
# Plot marginal distributions by drawing ticks along the x and y axes.
# This function is intended to complement other plots by showing the location of indecidual observations in an unobtrusive way.
sns.kdeplot(data= tips, x= 'total_bill')
sns.rugplot(data= tips, x= 'total_bill')
```

### Bivariate histogram
- A bivariate histogram bins the data within rectangles that tile the plot
- and then show the count of observations within each rectangle with the fill color

```python
sns.histplot(data= tips, x= 'total_bill', y= 'tip')
# sns.displot(data= tips, x= 'totla_bill', y= 'tip', kind= 'hist)    (same figure by using displot)
```

# Bivariate kdeplot
- A bivariate KDE plot smoothes the (x,y) observation with a 2D Gaussian function

```python
sns.kdeplot(data= tips, x= 'total_bill', y= 'tip')
```
