## Matplotlib Library
- For 2D Data Visulization
- Matplotlib is a multi-platform data visualization library built on NumPy arrays and designed to work with the border Scipy stack. It was introduced by John Hunter in the year 2002. One of the greatest benefits fo visualization is that it allows us visual access to huge amounts of data is easily digestible visuals.
#### Pyplot
- Most of the Matplotlib utilities lies under the pyplot submodule, and usually imported under alias:  import matplotlib.pyplot as plt
#### Types of chart using matplotlib
- Linear Plot
- Scatter Plot
- Bar Plot
- Stem Plot
- Step Plot
- Hist Plot
- Box Plot
- Pie Plot
- Fill_Between_plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Linear Graph

```python
x= [1,2,3,4]
y= [5,6,7,8]
c= 'r'         # Showing color
plt.plot(x,y,c)
plt.show()
```

```python
x= [5,2,9,4,7]
y= [10,5,8,4,2]
c= 'g'        
plt.plot(x,y,c)
plt.show()
```

```python
# Place marker

x= [5,2,9,4,7]
y= [10,5,8,4,2]
       
plt.plot(x,y, marker='o')
plt.show()
```

```python
# Saperate the lines of x and y 

x= [5,2,9,4,7]
y= [10,5,8,4,2]
 
plt.plot(x,y)
plt.plot(y, marker= '>')
plt.plot(x, marker= 'o')
plt.show()
```

#### Bar Graph

```python
x= [1,2,3,4]
y= [5,6,7,8]
c= ['r','g','b','y']   # select individualy colors for bar
plt.bar(x,y, color= c)
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
#plt.barh --> for horizontal bar graph
plt.bar(x,y, color='red', width= 0.4)   # bar color and width
plt.show()
```

```python
x= np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
#plt.barh --> for horizontal bar graph
plt.barh(x,y, color='red', height= 0.1)   # bar color and height
plt.show()
```

## Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x= [1,2,3,4,5,6]
y= [4,2,3,5,1,8]
plt.title('Scatter Plot')
plt.xlabel('Month')
plt.ylabel('Numbers')
c=['red','magenta', 'green','orange', 'blue', 'brown']
plt.scatter(x,y, color= c, s= 150) # (s=) is for size
plt.show()
```

### How to do changes into images

```python
# importing libraries

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

fname= r'Profile karan.png'

# opening image using PIL

image= Image.open(fname).convert('L')

# mapping image to color scale
plt.imshow(image, cmap= 'viridis_r')  # colors --> coolwarm, gray, rainbow, gist_heat_r, ocean_r, viridis_r
plt.show()
```

```python
fname= r'Profile karan.png'
image= Image.open(fname).convert('L')
plt.imshow(image, cmap='gray')
plt.title('karan singh')
plt.grid()   # for grid lines
plt.show()
```

### pie chart

```python
x= [10, 20,30,40]
y= ['english', 'hindi', 'maths', 'science']
c= ['yellow', 'magenta','brown','aqua']

plt.pie(x, labels= y, colors= c)
plt.legend()  # for label box
plt.show()
```
