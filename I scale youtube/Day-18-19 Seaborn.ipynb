## Seaborn Library
##### Seaborn is a library for making statistical graph in python. it builds on top of matplotlib and integrates closely with pandas data structures.
##### seaborn helps you explore and understand your data. its plotting functions operate on dataframe and arrays containing whole datasets and internally perform the mapping and statistical aggregation to produce informative plots.
##### Different categories of plot in seaborn:-
##### Plots are basically used for visualizing the relationship between variables. Those variables can be either completely numerical or a category like a group, class, or division. Seaborn divides the plot into the below categories:-
- Relational plots :- This plot is used to understand the relation between two variables.
- Categorical plots :- This plot deals with categorical variables and how they can be visualized.
- Distribution plots :- This plot is used for examining univariate and bivariate distributions.
- Regression plots :- The regression plots in Seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.
- Matrix plots :- A matrix plot is an array of scatterplots.
- Multi-plot grids :- It is a useful approach to draw multiple instances of the same data but different subsets of the dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var= [1,2,3,4,5,6,7,8]
var1= [8,7,6,5,4,3,2,1]
```

```python
# Graph by matplotlib library to see difference with seaborn library

plt.plot(var, var1)
plt.show()
```

```python
# same graph by seaborn library

var= [1,2,3,4,5,6,7,8]
var1= [8,7,6,5,4,3,2,1]

sns.lineplot(x=var, y=var1)
plt.show()
```

```python
var= [1,2,3,4,5,6,7,8]
var1= [8,7,6,5,4,3,2,1]

# by using pandas library and seaborn library
df= pd.DataFrame({"var":var, "var1":var1}) 
sns.lineplot(x= 'var', y= 'var1', data=df)
plt.show()
```

### Order Parameter

```python
order1= ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x= 'island', y= 'bill_length_mm', data= df1, hue= 'sex', order= order1)
plt.show()
```

```python
# Change hue order

order1= ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x= 'island', y= 'bill_length_mm', data= df1, hue= 'sex', 
            order= order1, hue_order= ['Female','Male'])
plt.show()
```

```python
# Default colors

order1= ['Dream','Torgersen','Biscoe']
sns.barplot(x= 'island', y= 'bill_length_mm', data= df1, color= 'pink')
plt.show()
```

```python
# Palette colors --> Accent, gist, prism, gist_earth

sns.barplot(x= 'island', y= 'bill_length_mm', data= df1, hue= 'sex', palette= 'gist_earth')
plt.show()
```

```python
# Saturation parameter for color grading

sns.barplot(x= 'island', y= 'bill_length_mm', data= df1, hue= 'sex', 
            palette= 'gist_earth', saturation= 0.3)
plt.show()
```
